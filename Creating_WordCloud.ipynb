```
#importing the required libraries and packages
import numpy as np
import pandas as pd

#reading the dataset
dataset = pd.read_csv('/content/drive/My Drive/recovery-news-data.csv')
dataset.head(10)
```

```
#forming separate datasets on reliable and unreliable sources of data
reliable_dataset = dataset[dataset['reliability'] == 1]
unreliable_dataset = dataset[dataset['reliability'] == 0]

#checking the length of both the datasets
print("Length of Reliable Dataset: {}".format(len(reliable_dataset)))
print("Length of Unreliable Dataset: {}".format(len(unreliable_dataset)))
```

```
#generating WordCloud of Reliable Dataset

#importing all necessary modules
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd

reliable_contents = ""
stopwords = set(STOPWORDS)

for text in reliable_dataset['body_text']:
  value = str(text)
  arr = value.split()
  
  #changing all the words to lowercase
  for i in range(len(arr)):
    arr[i] = arr[i].lower()

  reliable_contents += " ".join(arr) + " "

#forming the wordcloud
reliable_wordcloud = WordCloud(width = 800, height = 800, background_color ='white', stopwords = stopwords, min_font_size = 10).generate(reliable_contents)

#plotting the WordCloud image
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(reliable_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0)

plt.show()  
```

```
#generating WordCloud of Unreliable Dataset

#importing all necessary modules
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd

unreliable_contents = ""
stopwords = set(STOPWORDS)

for text in unreliable_dataset['body_text']:
  value = str(text)
  arr = value.split()
  
  #changing all the words to lowercase
  for i in range(len(arr)):
    arr[i] = arr[i].lower()

  unreliable_contents += " ".join(arr) + " "

#forming the wordcloud
unreliable_wordcloud = WordCloud(width = 800, height = 800, background_color ='white', stopwords = stopwords, min_font_size = 10).generate(unreliable_contents)

#plotting the WordCloud image
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(unreliable_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0)

plt.show()
```
